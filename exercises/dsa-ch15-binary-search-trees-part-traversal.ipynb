# 1. `TreeNode`, `Binary Tree` & `insert_node`: Setup

Introduced [previously](https://tonyjustdevs.github.io/blog/posts/computerscience/datastructuresandalgorithms/dsa-043-binary-search-trees-part-6.html).

```python
class TreeNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right


def insert_node(root_node: TreeNode, target: int):
    current_node = root_node
    if target == current_node.data:
        print(f"t[{target}]|c[{current_node.data}]: \t\tCannot insert duplicates")
        return  
    if target < current_node.data: # left
        if current_node.left: # go left-child: if exists
            print(f"t[{target}]|c[{current_node.data}]: \t\tCurrent[{current_node.data}] has ONE left-child : GO LEFT[{current_node.left.data}]")
            current_node=current_node.left
            return insert_node(current_node, target)
        else: # insert left-child: if not exist
            current_node.left = TreeNode(target)
            print(f"t[{target}]|c[{current_node.data}]: \t\tCurrent[{current_node.data}] has NO left-child: INSERTED LEFT CurrentLeft[{current_node.left.data}]")
            return current_node.left
    else:
        if current_node.right: # go right-child: if exists
            print(f"t[{target}]|c[{current_node.data}]: \t\tCurrent[{current_node.data}] has ONE right-child : GO right[{current_node.right.data}]")
            current_node=current_node.right
            return insert_node(current_node, target)
        else: # insert right-child: if not exist
            current_node.right = TreeNode(target)
            print(f"t[{target}]|c[{current_node.data}]: \t\tCurrent[{current_node.data}] has NO right-child: INSERTED right Currentright[{current_node.right.data}]")
    pass

def insert_node_clean(root_node: TreeNode, target: int):
    current_node = root_node
    if target == current_node.data:
        return  
    if target < current_node.data: # left
        if current_node.left: # go left-child: if exists
            current_node=current_node.left
            return insert_node_clean(current_node, target)
        else: # insert left-child: if not exist
            current_node.left = TreeNode(target)
            return current_node.left
    else:
        if current_node.right: # go right-child: if exists
            current_node=current_node.right
            return insert_node_clean(current_node, target)
        else: # insert right-child: if not exist
            current_node.right = TreeNode(target)
    pass

# create a basic tree
#             50
#      25           75
#  10     33     56     89  
# 4  11 30  40 52  61 82  95  
def insert_node_list(root_node:TreeNode, node_list: list[int], show_outputs:bool=True):
    for node in node_list: 
        if show_outputs:
            insert_node(root_node, node) 
        else:
            insert_node_clean(root_node, node) 

def create_tree_from_root_node(root_node = TreeNode(50),node_list=[25,75,10,33,56,89,4,11,30,40,52,61,82,95]):
    insert_node_list(root_node, node_list,show_outputs=False)
    
```

# 3. `delete_node`: 0 or 1 Child Nodes Only  

```python
def find_successor_node(target_node): # t has 2 children
    print(f"t[{target_node.data}]t.r[{target_node.right.data}]: went right, current s_node[{target_node.right.data}]")
    successor_node = target_node.right
        #             [50x]
        #      25           [75s]
        #  10     33     56     89  
        # 4  11 30  40 [52]  61 82  95
    
    if not successor_node.left:
        target_node = successor_node
        target_node.right = successor_node.right
        print(f"t[{target_node.data}]s_node[{successor_node.data}]: has no left_child, replace target with s_node, and target.right with s_node.right")
        return 
        #             [50x]
        #      25           [75s] # 56 doesnt exist, [75s] replaces [50x]j
        #                       89
    
    while successor_node.left:
        parent_node = successor_node      
        successor_node = successor_node.left
        print(f"t[{target_node.data}].s_node.l[{successor_node.data}]: [{parent_node.data}] went left s_node[{successor_node.data}]")
    
    print(f"t[{target_node.data}].s_node[{successor_node.data}]: arrived at left_most child s_node[{successor_node.data}]")

    if successor_node.right:
        print(f"t[{target_node.data}].s_node[{successor_node.data}]: s_node[{successor_node.data}] has a right_child[{successor_node.right}]")
        print(f"t[{target_node.data}].s_node[{successor_node.data}]: s_node[{successor_node.data}] to replace target[{target_node.data}], s_node.right[{successor_node.right.data}] replaces s_node[{successor_node.data}], s_node_parent[{parent_node.data}].left[{parent_node.left.data}] to be s_node.right[{successor_node.right.data}]")
        successor_node = successor_node.right # [55] exists (previously [51])
        parent_node.left = successor_node #[56].left = [55] (previously [52] aka s_node)
        #             [x50]
        #      25           75
        #  10     33      56     89  
        # 4  11 30  40 [52]  61 82  95
        #                [55]
    else:
        print(f"t[{target_node.data}].s_node[{successor_node.data}]: s_node[{successor_node.data}] has no right child")
        print(f"t[{target_node.data}].s_node[{successor_node.data}]: s_node[{successor_node.data}] to replace target[{target_node.data}], s_node_parent[{parent_node.data}]l[{parent_node.left.data}] to be: None")
        parent_node.left = None # 55 doesnt exist
        #             [x50]
        #      25           75
        #  10     33     56     89  
        # 4  11 30  40 [52]  61 82  95
            
    target_node = successor_node 

    return target_node
```

```python
def delete_node(root_node: TreeNode, target: int):
    current_node = root_node 
    parent_node  = None
    target_node  = None
    
    while current_node:
        if target == current_node.data:
            target_node = current_node
            print(f"1: target found: t[{target}]c[{current_node.data}]")
            break
        elif target < current_node.data:
            parent_node = current_node
            current_node = current_node.left
        elif target > current_node.data:
            parent_node = current_node
            current_node = current_node.right # None if
        else:
            print("This shouldnt print")
    
    if not target_node: # root has no parents
        print(f"2: target not found: t[{target}]c[{None}]p[{parent_node.data}]")

    if target_node.left and target_node.right:
        target = find_successor_node(target_node)
        return
    else:
        target_child_node = (target_node.left or target_node.right)
        
        
        if not parent_node: # root
            print(f"3: target: t[{target}]: is root replace with t[{target_node.data}]-child)] \
                and its children whether None or not")
                #   t[{target_node.data}]l[{target_node.left.data}] by tcl[{target_child_node.left.data}] \
                #   t[{target_node.data}]r[{target_node.right.data}] by tcr[{target_child_node.right.data}]")   
            target_node = target_child_node
            target_node.left  = target_child_node.left
            target_node.right = target_child_node.right
        elif parent_node.left == target_node:
            print(f"4: target: t[{target}]: p[{parent_node.data}] by t[{target_node.data}]'s child")
            # print(f"4: target: t[{target}]: p[{parent_node.data}]l[{parent_node.left.data}] to be replaced by t[{target_node.data}]c[{target_child_node.data}]")
            parent_node.left = target_child_node # tcn is None if leaf
        elif parent_node.right == target_node:
            print(f"5: target: t[{target}]: p[{parent_node.data}]r[{parent_node.right.data}] to be replaced by t[{target_node.data}]'s child")
            # print(f"5: target: t[{target}]: p[{parent_node.data}]r[{parent_node.right.data}] to be replaced by t[{target_node.data}]c[{target_child_node.data}]")
            parent_node.right = target_child_node  # tcn is None if leaf
        
    if target_child_node:
        print(f"target_child_node: [{target_child_node.data}]")
    else:
        print(f"{target} has no chilren")
    return target
        
```
